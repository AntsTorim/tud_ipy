```python
import kernel_fca_oo as krn
import pandas as pd

hiiumaa = pd.read_csv("TTTH_kardlaga.csv", encoding="cp1257", index_col=0) # Baltic languages encoding
#hiiumaa = pd.read_csv("TTTH_kardlaga.csv", encoding="cp1257") # Baltic languages encoding
hiiumaa.describe()
```

```python
#for dt in hiiumaa.dtypes: print(dt)
#hiiumaa = hiiumaa.loc[:, 1:]
#hiiumaa=hiiumaa.loc[:, 'lapsi10':
#print(hiiumaa.columns)
new_index = list(range(len(hiiumaa.index)))
#print(new_index)
#print(hiiumaa.index[hiiumaa.index.duplicated()])
#hiiumaa.reindex(new_index)
ks = krn.FCASystemDF(hiiumaa)
cc = ks.conceptcover(uncovered=0.2)
print("\n\n\nConcepts:", len(cc), "\n")
for (e, i) in cc:
    if len(e) > 9:
        print(len(e), i)
    else:
        print(e, i)

ks = krn.FCASystemDF(hiiumaa.transpose())
cc = ks.conceptcover(uncovered=0.2)
print("\n\n\nTransposed concepts:", len(cc), "\n")
for (e, i) in cc:
    if len(i) > 9:
        print(e, len(i))
    else:
        print(e, i)
```
