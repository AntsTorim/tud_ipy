```python
c_i = range(0, 100)
KATALOOG = "..\\"

fns = ["cmp-bin-zoo.csv",
       "priv2.csv",
       "cmp-bin-iapb09.csv",
       "cmp-bin-STT_majandus.csv",
       "cmp-bin-TTTH_kardlaga.csv",
       "cmp-bin-grocery.csv", 
       "cmp-bin-ecommerce-sampled.csv", 
       "cmp-bin-house-votes-84.csv",
       "cmp-bin-instacart-600-sample.csv",
       "cmp-bin-student-gradings.csv",
       "cmp-bin-taro-1.csv",
       "cmp-bin-taro-2.csv",
       "cmp-bin-taro-3.csv",
       "cmp-bin-taro-4.csv"]

for fn in fns:
    print("\nFile:", fn)
    data = pd.read_csv(KATALOOG+fn, sep=',', index_col=0, encoding='latin1')
    result_df = pd.DataFrame()
    #systems = {"MS": FCASystemDF(data), "FL": FreqLexiSeriateSystem(data)}
    systems = {"CL": ConfLexiSeriateSystem(data), "FL": FreqLexiSeriateSystem(data)}
    ones = data.sum().sum()
    for s_name, system in systems.items():
        start = timeit.default_timer()
        #_, uc =  system.conceptchaincover_v2(uncovered=0.0, max_cc=100)
        _, uc =  system.conceptchaincover(uncovered=0.0, max_cc=100)
        time = timeit.default_timer() - start
        if len(uc) < len(c_i):
            uc += [0] * (len(c_i) - len(uc))  
        result_df[s_name] = [int((1-x)*ones) for x in uc[:len(c_i)]]
    #[uc for (_, uc) in zip(system.conceptchaincover())]
    display(result_df)
    result_df.to_csv(path_or_buf=KATALOOG+"dres-"+fn)
```
