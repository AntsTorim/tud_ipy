```python
# Sihi datagen
for i in range(1, 10):
    dat, elo, elo_prob = generate_data_by_input(generate_samples_amount = 100, 
                                            generate_tests_amount = 20, 
                                            aspects_amount = i,
                                            test_aspect_ratios = '')
    cl_s = ConfLexiSeriateSystem(pd.DataFrame(dat))
    ccc, uc = cl_s.conceptchaincover(uncovered=0.1, max_cc=20)
    print("True aspects", i)
    print("Uncovered", uc)
    uc = [1.0] + uc
    plt.plot(uc, color="0.%i" % i, label=str(i))
    plt.legend()
```

```python
# Minu datagen
for i in range(1, 10):
    dat, elo, elo_prob = make_test_results(n_samples = 100, 
                                           n_tests = 20, 
                                           n_aspects = i)
    cl_s = ConfLexiSeriateSystem(pd.DataFrame(dat))
    ccc, uc = cl_s.conceptchaincover(uncovered=0.1, max_cc=20)
    print("True aspects", i)
    print("Uncovered", uc)
    uc = [1.0] + uc
    plt.plot(uc, color="0.%i" % i, label=str(i))
    plt.legend()
```
